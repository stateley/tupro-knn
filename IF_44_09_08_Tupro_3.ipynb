```python
#Normalize Function

column = ['x1', 'x2', 'x3']
def normalize(data, column, factor=1):
    result = data.copy()
    for col in column:
        max_value = data[col].max()
        min_value = data[col].min()
        result[col] = ((data[col] - min_value) / (max_value - min_value))*factor
    return result
```

```python
#Slicing data_train into data_val

data_train = data_train1.iloc[:270]
data_val = data_train1.iloc[270:296]
```

```python
#Data train normalization process

cols_to_norm = ['x1','x2', 'x3']
data_train[cols_to_norm] = data_train[cols_to_norm].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
data_train.head()
```

```python
#Data validation normalization process

cols_to_norm = ['x1','x2', 'x3']
data_val[cols_to_norm] = data_val[cols_to_norm].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
data_val.head()
```
